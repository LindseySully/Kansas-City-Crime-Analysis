## Initial Visualization

```python
import pandas as pd
csv_file_path = 'Data/Data_Sets/KCPD_5_Year_Analysis_Cleaned.csv'
df = pd.read_csv(csv_file_path)
```

#### General Offense Categorization Count

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Setting figure size
plt.figure(figsize=(12, 10))  # You can adjust these numbers as needed

# Setting font scale
sns.set(font_scale=1.2)  # Adjusts the size of the labels, title

# Creating Count Plot
sns.countplot(data=df, y='General _Offense_Categorization', order = df['General _Offense_Categorization'].value_counts().index)
plt.title('Counts of General Offense Categories')
plt.xlabel('Count')
plt.ylabel('Offense Category')

plt.tight_layout()

plt.show()
```

#### Distribution of Age

```python
sns.histplot(data=df, x='Age', bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# filter out null values
df_filtered = df[df['Age_Range'].notnull() & (df['Age_Range'] != 'Null')]

sns.histplot(data=df_filtered, x='Age_Range', bins=30)
plt.title('Age Range Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

#### Heatmap of Crime Type and Age Range

```python

age_crime_crosstab = pd.crosstab(df['Age_Range'], df['Type_of_Crime'])
sns.heatmap(age_crime_crosstab, annot=True, fmt='d', cmap='coolwarm')
plt.title('Heatmap of Crime Type and Age Range')
plt.xlabel('Type of Crime')
plt.ylabel('Age Range')
plt.show()
```

#### UCR Offense Classification Proportions

```python
ucr_counts = df['UCR_Offense_Classification'].value_counts()
plt.pie(ucr_counts, labels=ucr_counts.index, startangle=90, counterclock=False, autopct='%1.1f%%')
plt.title('UCR Offense Classification Proportions')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

#### NIBRS Offense Group Proportions

```python
nibrs_counts = df['NIBRS Offense Group'].value_counts()
plt.pie(nibrs_counts, labels=nibrs_counts.index, startangle=90, counterclock=False, autopct='%1.1f%%')
plt.title('NIBRS Offense Group Proportions')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

### Time Analysis

#### Yearly Crime Counts in Kansas City

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert Reported_Date to datetime
df['Reported_Date'] = pd.to_datetime(df['Reported_Date'])

# Resample to get yearly counts
yearly_crimes = df.resample('Y', on='Reported_Date').size()

# Plot
yearly_crimes.plot(title='Yearly Crime Counts')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.show()
```

#### Crime Counts by Day of the Week

```python
# Organize Days of the Week
days = ["Sunday","Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]
day_type = pd.CategoricalDtype(categories=days, ordered=True)

# Convert Day_of_Week to ordered categorical data
df['Day_of_Week'] = df['Day_of_Week'].astype(day_type)

# Count values and sort by the categorical data
weekday_crimes = df['Day_of_Week'].value_counts().sort_index()

# Plot Bar Chart
plt.figure(figsize=(10,5))
weekday_crimes.plot(kind='bar', color='skyblue', edgecolor = 'black',title='Crime Counts by Day of the Week')
plt.title('Crime Counts by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle ='--', linewidth=0.7,alpha=0.7) # Add grid lines for readability
plt.tight_layout()
plt.show()
```

#### Heatmap of Crime Frequency by Hour and Day

```python
# Convert Reported_Time to datetime and extract hour
df['Reported_Hour'] = pd.to_datetime(df['Reported_Time']).dt.hour

# Sort Days of the Week
df['Day_of_Week'] = df['Day_of_Week'].astype(day_type)

# Create a crosstab heatmap
heatmap_data = pd.crosstab(df['Reported_Hour'], df['Day_of_Week'])

import seaborn as sns

# Plot
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap='viridis')
plt.title('Heatmap of Crime Frequency by Hour and Day')
plt.show()
```

#### Boxplot of Crimes by Hour of Day

```python
# Plot Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Reported_Hour', data=df)
plt.title('Boxplot of Crimes by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Crime Count')
plt.show()
```

### Correlation

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Month_Num'] = le.fit_transform(df['Month'])

# Select temporal features
temporal_features = df[['Year', 'Quarter', 'Month_Num', 'Reported_Hour']]

# Calculate correlation matrix
correlation_matrix = temporal_features.corr()

# Plot
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix of Temporal Features')
plt.show()
```
